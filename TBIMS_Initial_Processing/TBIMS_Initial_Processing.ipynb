```python
data_path = r'C:\Users\elyas\Desktop\Dissertation_PhD_Proposal\Dataset_Expansion\TBIMSP\TBIMSPublic.2024-11-01\Data'

form1_file = os.path.join(data_path, 'Form1.csv')
form2_file = os.path.join(data_path, 'Form2.csv')
tbiid_file = os.path.join(data_path, 'TBIID.csv')

print("Loading files...")
form1 = pd.read_csv(form1_file, low_memory=False)
form2 = pd.read_csv(form2_file, low_memory=False)
tbiid = pd.read_csv(tbiid_file, low_memory=False)
print("Files loaded successfully.")

print("Merging Form1 and Form2 on Mod1Id...")
merged = pd.merge(form1, form2, on='Mod1Id', how='left', suffixes=('_Form1', '_Form2'))

print("Merging with TBIID on Mod1Id and Mod2Id...")
merged = pd.merge(merged, tbiid, on=['Mod1Id', 'Mod2Id'], how='left')

output_path = os.path.join(data_path, 'TBIMS_Merged.csv')
merged.to_csv(output_path, index=False)

print("\n‚úÖ Merge Complete!")
print(f"Saved merged file to: {output_path}")
print(f"Rows: {len(merged)} | Columns: {len(merged.columns)}")
```

```python
import numpy as np

cleaned_file_path = r'C:\Users\elyas\Desktop\Dissertation_PhD_Proposal\Dissertation\Cleaning_Up\TBIMS_Merged_Missing_Values_cleaned.csv'  
df_cleaned = pd.read_csv(cleaned_file_path, low_memory=False)

special_codes_to_nan = ['66', '77', '88', '99', '99999', '8888', '09/09/9999']
df_cleaned.replace(special_codes_to_nan, np.nan, inplace=True)
```

```python
#Removing columns where >90% of the values are 88888 (e.g., ZIP code fields) as they are not useful for modeling.
```

```python
cols_to_drop = []
for col in df_cleaned.columns:
    if (df_cleaned[col].astype(str) == '88888').mean() > 0.9:
        cols_to_drop.append(col)

df_cleaned.drop(columns=cols_to_drop, inplace=True)
print("Dropped columns:", cols_to_drop)
```

```python
#Adding a new column DeathIndicator to flag if any row contains the death-related code 44444. (1 = death-related, 0 = not)
```

```python
df_cleaned['DeathIndicator'] = df_cleaned.apply(lambda row: 1 if '44444' in row.values else 0, axis=1)
```

```python
#Checking Missingness ‚Üí Dropping Columns with High Missingness (>50‚Äì60%)
```

```python
missingness = df_cleaned.isnull().mean().sort_values(ascending=False) * 100

columns_to_drop = missingness[missingness > 60].index.tolist()

print("Columns to drop (missing >60%):")
print(columns_to_drop)

df_cleaned = df_cleaned.drop(columns=columns_to_drop)

print(f"\n‚úÖ Dropped {len(columns_to_drop)} columns with high missingness.")
```
